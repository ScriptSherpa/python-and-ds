```python
# Bar graph
# bar()
plt.bar([1,2,3,4],[5,6,7,8],label='data1')
plt.bar([6,9,5,7],[7,8,1,2],label='data2')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Graph')
plt.show()
```

```python
# Horizontal Bar graph
#BARH()

products=['computers','laptops','mouse','keyboards']
quantities=[234,456,678,890]
plt.barh(products,quantities,label='products')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Horizontal Bar Graph')
plt.show()
```

```python
#scatter plot
# scatter

x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]
plt.scatter(x,y,label='skitscat',color='r',s=25,marker='*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
# pie plot
# pie()

slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,
        explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.legend()
```
