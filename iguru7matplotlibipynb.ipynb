```python
# Data Visualization
# Matplotlib
# https://matplotlib.org/

# Data visualization is the representation of data through use of common graphics,
# such as charts, plots, infographics and even animations.
# The primary goal of data visualization is to make data more accessible and
# easier to interpret, allowing users to identify patterns, trends, and outliers quickly.

# Matplotlib is a comprehensive library for creating static, animated,
# and interactive visualizations in Python.
# Matplotlib makes easy things easy and hard things possible.

# You can generate plots, histograms, power spectra, bar charts, errorcharts,
# scatterplots, etc., with just a few lines of code.

# The matplotlib library contains several modules. The matplotlib.pyplot module is
# designed to create colorful plots.
```

```python
import numpy as np
import pandas as pd
import matplotlib
print(matplotlib.__version__)
import matplotlib.pyplot as plt
```

```python
# Line graph
# plot(x,y)---> draws a line from a point to point
plt.plot([1,2,3],[4,5,1])
#show()---> used to display the grapgh
plt.show()
```

```python
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.style.use('ggplot')
plt.show()
```

```python
# Adding styles in graphs
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
#matplotlib.pyplot.legend — Matplotlib 3.9.3 documentation


```

```python
from matplotlib import style
style.use('ggplot')
x1=[5,8,10]
x2=[6,9,11]
y1=[12,16,6]
y2=[6,10,14]
plt.plot(x1,y1,color='green',linewidth=3,label='product1',linestyle='dashed')
plt.plot(x2,y2,color='red',linewidth=3,label='product2',linestyle='dotted') # dotted/dashed /solid
plt.title("Line Graph using style")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(loc=2)  # Helps to display labels
plt.show()
```
