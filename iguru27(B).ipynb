```python
#https://www.anaconda.com/download/success

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(sns.__version__)
```

```python
# built in data sets of seborn
sns.get_dataset_names()
```

```python
dataset= sns.load_dataset('iris')
dataset
```

```python
# Line Graph --> lineplot()
sns.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})
sns.lineplot(x='sepal_length', y='sepal_width', data=dataset)
plt.title('Line Graph in Seaborn')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
pokemon =pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv")
pokemon
```

```python
# Scatter plot
sns.scatterplot(x='Attack', y='Defense', hue='Generation',data=pokemon)
plt.title('Scatter Plot in Seaborn')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```

```python
pokemon.Generation.value_counts()
```

```python
colors=['r','y','b','g','c','m']
sns.scatterplot(x='Attack', y='Defense', hue='Generation',palette=colors,data=pokemon)
plt.title('Scatter Plot in Seaborn')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```

```python
#count plot
# generate bar graph

plt.figure(figsize=(10,5))
sns.countplot(x='Type 1',data=pokemon)
plt.title('Count Plot in Seaborn')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Type 1',palette=pkmn_type_colors,data=pokemon)
plt.title('Count Plot in Seaborn')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.show()
```
