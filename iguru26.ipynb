```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


```

```python
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv')
data
```

```python
# Generating graph using imported dataset csv file

plt.scatter(data.Hours_Studied,data.Test_Grade,color='red')
plt.plot(data.Hours_Studied,data.Test_Grade,color='green',label='Actual Data Values ')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

```python
# xticks () and yticks ()
from numpy import random
import matplotlib.pyplot as plt
import matplotlib

x = random.randint(100, size=(10))
y = random.randint(60, size=(10))
plt.plot(x, y, color = 'hotpink')
plt.xticks(np.arange(0, 105, 5))
plt.yticks(np.arange(0, 65, 5))
plt.title("Xticks and Yticks ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
